```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
file_path = 'fifa_eda_stats.csv'
dataFoot = pd.read_csv(file_path)

# Vista previa de los datos
print("Vista previa de los datos:")
print(dataFoot.head())

```

```python
# Resumen estadístico
print("\nResumen estadístico:")
print(dataFoot.describe())

```

```python
# Información general
print("\nInformación general:")
print(dataFoot.info())
```

```python
# Análisis gráfico

# Histograma de las variables numéricas
columnas_numericas = dataFoot.select_dtypes(include=['int64', 'float64']).columns

for col in columnas_numericas:
    plt.figure(figsize=(10, 6))
    dataFoot[col].hist(bins=30)
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()
```

```python
# Matriz de correlación
corr_matrix = dataFoot[columnas_numericas].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 6})
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Gráficas de dispersión para algunas variables
import seaborn as sns

importantes = ['Overall', 'Potential']
sns.pairplot(dataFoot[importantes])
#plt.suptitle('Gráficas de Dispersión para Overall y Potential', y=1.02)
plt.show()

```

```python
# Boxplots para ver la distribución de algunas variables por categorías
categorical_columns = dataFoot.select_dtypes(include=['object']).columns

for col in categorical_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=col, y='Overall', data=dataFoot)  # 'Overall' puede ser reemplazado por otra variable de interés
    plt.title(f'Boxplot de Overall por {col}')
    plt.xticks(rotation=90)
    plt.show()
```
